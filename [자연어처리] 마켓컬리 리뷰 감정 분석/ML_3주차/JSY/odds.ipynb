#### Odds_Ratio

logistic regression: 이진분류를 수행할 때 사용되는 모델로, 데이터 샘플을 양성(1) or 음성(0) class 중, 어디에 속하는지 예측한다. 각 속성(featrue)들의 계수(log_odds_ratio)를 구한 후, 시그모이드 함수를 적용하여 실제로 데이터가 해당 클랫에 속할 확률을 구한다.

---

Step1 <br>
$p \in [0,1]$ : 1에 속할 확률

---

Step2 <br>
$\text{odds} = \cfrac{p}{1-p} \in [0,\infty]$ : 1에 속할 확률이 0에 속할 확률에 비해 몇 배 높은지(사건이 일어날 확률이 일어나지 않을 확률에 몇배인지)

---

Step3 <br>
$\text{logit} = \text{log(odds)} = \log\left(\cfrac{p}{1-p}\right) = \beta_{0} + \beta_{1}x_{1} + \cdots + \beta_{k}x_{k} \in [-\infty,\infty] $ <br>
$\beta_{0} + \beta_{1}(x_{1} + 1) + \cdots + \beta_{k}x_{k} = \beta_{0} + \beta_{1}x_{1} + \cdots + \beta_{k}x_{k} + \beta_{1}$

$\beta_{k}$(계수)는 feature $x_{k}$값이 1증가 했을 때, 1로 분류될 확률의 증가량(logit의 증가량) 



---

Step4 <br>
$\text{odds} = \cfrac{p}{1-p} = e^{\beta_{0} + \beta_{1}x_{1} + \cdots + \beta_{k}x_{k}} \in [0, \infty] $ <br>

$p = \cfrac{1}{1 + e^{-(\beta_{0} + \beta_{1}x_{1} + \cdots + \beta_{k}x_{k})}}$

---

Step5 <br>
odds_ratio: 나머지 feature들을 모두 고정시킨 상태에서 한 feature를 1단위 증가시켰을 때, 변화하는 odds의 비율이다. <br>

$\text{odds_ratio} = \cfrac{\text{odds}(x_{1}+1, x_{2}, \cdots, x_{k})}{\text{odds}(x_{1}, x_{2}, \cdots, x_{k})} = \cfrac{e^{\beta_{0} + \beta_{1}(x_{1}+1) + \cdots + \beta_{k}x_{k}}}{e^{\beta_{0} + \beta_{1}x_{1} + \cdots + \beta_{k}x_{k}}} = e^{\beta_{1}}$ <br>

즉, $(x_{1}, ..., x_{k})$일 때 odds가 높은지, 즉, $(x_{1}+1, ..., x_{k})$일 때 odds가 높은지를 수치적으로 표현한다.

---

Step6 <br>

$\text{odds_ratio} = e^{\beta_{1}} \geq 1 \, \Rightarrow \, \text{odds}(x_{1}+1, x_{2}, \cdots, x_{k}) \geq \text{odds}(x_{1}, x_{2}, \cdots, x_{k})$ <br>
$\text{odds_ratio} = e^{\beta_{1}} \leq 1 \, \Rightarrow \, \text{odds}(x_{1}+1, x_{2}, \cdots, x_{k}) \leq \text{odds}(x_{1}, x_{2}, \cdots, x_{k})$ <br>

$\beta_{k} \geq 0 \Rightarrow x_{k} \, \text{feature는 1일 확률과 양의 관계}$ <br>
$\beta_{k} < 0 \Rightarrow x_{k} \, \text{feature는 1일 확률과 음의 관계}$


---

Step7 <br>

$\text{odds_ratio} = \exp(\text{회귀계수})$ <br>


---

Summary : Logistic Regression

$h(x) = \text{Prob}(\text{1에 속할 확률}) = \cfrac{1}{1 + e^{-x}} \,\, $ where $x = \beta_{0} + \beta_{1}x_{1} + \cdots + \beta_{k}x_{k}$ <br>
독립변수들 $(x_{1}, ..., x_{k})$에 의해서 $x$값이 변화하고 $x$는 최종적으로 Prob에 영향을 준다. <br>
즉, 어떤 요인(키워드)들이 리뷰의 긍정, 부정에 영향을 미치는 것인지, 얼만큼 영향을 끼치는 가를 알 수 있다.
