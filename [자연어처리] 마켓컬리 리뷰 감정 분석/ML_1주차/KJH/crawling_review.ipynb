### soynlp 이모티콘, 반복단어 제거

```python
df_reviwe_test = reviwe_df.head().copy()
```

```python
df_reviwe_alltest = reviwe_df.copy()
```

```python
# soynlp 다운, emoji 다운
#!pip install soynlp
#!pip install emoji
```

```python
import re
import emoji
from soynlp.normalizer import emoticon_normalize, repeat_normalize
```

```python
pattern = re.compile(f'[^ .,?!/@$%~％·∼()\x00-\x7Fㄱ-ㅣ가-힣]+')
url_pattern = re.compile(
    r'https?:\/\/(www\.)?[-a-zA-Z0-9@:%._\+~#=]{1,256}\.[a-zA-Z0-9()]{1,6}\b([-a-zA-Z0-9()@:%_\+.~#?&//=]*)')

def clean(x): 
    x = pattern.sub(' ', x)
    x = emoji.replace_emoji(x, replace='') #emoji 삭제
    x = url_pattern.sub('', x)
    x = x.strip()
    x = emoticon_normalize(x, num_repeats=2)
    x = repeat_normalize(x, num_repeats=2)
    return x
```

```python
clearing_by_soynlp = []
for i in range(len(df_reviwe_alltest)):
    clean_sentence = clean(df_reviwe_alltest['리뷰'][i])
    clearing_by_soynlp.append(clean_sentence)
clearing_by_soynlp
```

```python
df_reviwe_alltest['리뷰'] = clearing_by_soynlp
df_reviwe_alltest.head()
```

### py-hanspell로 띄어쓰기, 맞춤법 정제

```python
# 띄어쓰기, 맞춤법 py-hanspell 다운로드
#!pip install git+https://github.com/ssut/py-hanspell.git
```

```python
from hanspell import spell_checker
```

```python
# 맞춤법, 띄어쓰기 교정 **오류가 떠서 예외처리 사용
clearing_by_hanspell = []
for i in range(len(df_reviwe_alltest)):
    try:
        sent = df_reviwe_alltest['리뷰'][i]
        spelled_sent = spell_checker.check(sent)
        clearing_by_hanspell.append(spelled_sent.checked)
    except:
        clearing_by_hanspell.append(df_reviwe_alltest['리뷰'][i])
clearing_by_hanspell
```

```python
df_reviwe_alltest['리뷰'] = clearing_by_soynlp
df_reviwe_alltest.head()
```
