```python
from hanspell import spell_checker
```

```python
import time
start = time.time()

renew_hanspell = []

for review in df_review__stop1['review']:
    
    try:
        tmp = spell_checker.check(review)
        renew_review = tmp.checked

        renew_hanspell.append(renew_review)
    except:
        renew_hanspell.append(review)
        print(df_review__stop1[df_review__stop1['review'] == review].index[0])
        
print(time.time() - start)
```

```python
for i in renew_hanspell:
    print(i)
```
