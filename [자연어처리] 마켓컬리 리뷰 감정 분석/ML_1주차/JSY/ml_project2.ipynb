```python
from soynlp.utils import DoublespaceLineCorpus
from soynlp.noun import LRNounExtractor_v2
from soynlp.word import WordExtractor
```

```python
df_review = pd.read_csv('/Users/seongeunjeong/Documents/ds_study/ml_project/review_all_hanspell.csv')
df_review = df_review[['date', 'review', 'rating', 'label']]
df_review
```

```python
#review들을 하나의 문서로 합친다.
reviews_1000 = '. '.join(df_review['review'][:1000])  
reviews_1000
```

```python
open("reviews.txt", 'w')
```

```python
#DoublespaceLineCorpus: text 파일 path를 입력해야 한다.
corpus = DoublespaceLineCorpus('reviews.txt', iter_sent=True)
noun_extractor = LRNounExtractor_v2()
nouns = noun_extractor.train_extract(corpus)
```

```python
nouns
```

```python
#_compounds_components: 복합명사의 components가 저장되어 있다.
list(noun_extractor._compounds_components.items())
```

```python
#lrgraph.get_l: L-R 구조의 L parts을 확인한다.
noun_extractor.lrgraph.get_l('도')
```

```python
#lrgraph.get_R: L-R 구조의 R parts을 확인한다.
noun_extractor.lrgraph.get_r('좋')
```

```python
#WordExtractor의 parameter
word_extractor = WordExtractor(
    min_frequency=20,
    min_cohesion_forward=0.05, 
    min_right_branching_entropy=0.0
)

word_extractor.train(corpus)
words = word_extractor.extract()
```

```python
words
```

```python
import math

def word_score(score):
    return (score.cohesion_forward * math.exp(score.right_branching_entropy))

print('단어   (빈도수, cohesion, branching entropy)\n')
for word, score in sorted(words.items(), key=lambda x:word_score(x[1]), reverse=True)[:30]:
    print('%s     (%d, %.3f, %.3f)' % (
            word, 
            score.leftside_frequency, 
            score.cohesion_forward,
            score.right_branching_entropy
            )
         )
```

```python
#cohesion_scores
cohesion_scores = word_extractor.all_cohesion_scores()
cohesion_scores['신선']
```

```python
#branching_entropy
branching_entropy = word_extractor.all_branching_entropy()
branching_entropy['신선'] # (left_branching_entropy, right_branching_entropy)
```

```python
#accessor_variety
accessor_variety = word_extractor.all_accessor_variety()
accessor_variety['신선'] # (left_accessor_variety, right_accessor_variety)
```

```python
df_review_tmp = df_review.head(1000)
df_review_tmp
```

```python
#okt
okt_tokenizer = Okt()
```

```python
#WordExtractor의 parameter
word_extractor = WordExtractor(
    min_frequency=10,
    min_cohesion_forward=0.05, 
    min_right_branching_entropy=0.0
)

word_extractor.train(corpus)
words = word_extractor.extract()


#LTokenizer based on cohension_score

cohesion_score = {word:score.cohesion_forward for word, score in words.items()}
l_tokenizer = LTokenizer(scores=cohesion_score)
```

```python
okt = []
l = []

for review in df_review_tmp['review']:
    okt.append(okt_tokenizer.morphs(review))
    l.append(l_tokenizer.tokenize(review))
```

```python
#Okt
for i in okt:
    print(i)
```

```python
#LTokenizer
for i in l:
    print(i)
```
