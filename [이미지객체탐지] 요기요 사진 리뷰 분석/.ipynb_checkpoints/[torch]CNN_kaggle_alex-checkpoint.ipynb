# Food Torch CNN/AlexNet

```python
import os
import random
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")

from sklearn.model_selection import train_test_split

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader, random_split
from torchvision import datasets, transforms, models
from torchvision.utils import make_grid
```

```python
transform=transforms.Compose([
        transforms.RandomRotation(10),      # rotate +/- 10 degrees
        transforms.RandomHorizontalFlip(),  # reverse 50% of images
        transforms.Resize(224),             # resize shortest side to 224 pixels
        transforms.CenterCrop(224),         # crop longest side to 224 pixels at center
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406],
                             [0.229, 0.224, 0.225])
])
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
device
```

## Train and Test data Setting

```python
dataset=datasets.ImageFolder(root=("./imgs_data"),transform=transform)

class_names=dataset.classes
print(len(class_names))
print(class_names)
```

```python
torch.manual_seed(10)
test_size = len(dataset)//5
train_size = len(dataset) - test_size

train_ds, test_ds = random_split(dataset, [train_size, test_size])
len(train_ds), len(test_ds)
```

```python
batch_size = 10

train_loader=DataLoader(train_ds,batch_size=batch_size,shuffle=True)
test_loader=DataLoader(test_ds,batch_size=batch_size)
```

## Training function

```python
def Do_training(model, epochs, verbose_epoch_term=2):
    
    import time
    start_time=time.time()

    train_losses=[]; test_losses=[]; train_correct=[]; test_correct=[]; train_acc=[]; test_acc=[];

    len_batch = len(train_ds)//batch_size;
    
    print(f">>>> {model.__class__.__name__} Model fit start!!!"); print("  ")  
    for epoch in range(epochs):

        trn_corr=0; tst_corr=0;

        for b, (X_train,y_train) in enumerate(train_loader):
            b+=1
            X_train, y_train = X_train.to(device), y_train.to(device)

            y_pred=model(X_train)
            loss=criterion(y_pred,y_train)

            #true predictions
            predicted=torch.max(y_pred.data,1)[1]
            batch_corr=(predicted==y_train).sum()
            trn_corr+=batch_corr

            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            
            term = len(train_loader)//verbose_epoch_term
            if b%term==0:    # batch_size = 10, train_ds = 720 -> each batch : 72  ==> 72/36 =2번 출력
                print(f'epoch: {epoch+1:2}  batch: [{batch_size*b:4}/{len(train_ds)}]  loss: {loss.item():10.8f}  accuracy: {trn_corr.item()*100/(batch_size*b):7.3f}%')

        loss=loss.cpu().detach().numpy()
        train_losses.append(loss)
        train_correct.append(trn_corr)

        #-------- each Epoch accuracy -------------
        # accuracy = 100 * correct / len(trainset)
        # trainset, not train_loader
        # probably x in your case

        epoch_acc = trn_corr.item()/len(train_ds)
        train_acc.append(epoch_acc)
        print('  ')
        print(f"Epoch: {epoch+1:2}  Accuracy :{epoch_acc*100:7.3f}%" ) 
        print('-'*66)    

        with torch.no_grad():
            for b, (X_test,y_test) in enumerate(test_loader):
                X_test, y_test = X_test.to(device), y_test.to(device)

                y_val=model(X_test)
                loss=criterion(y_val,y_test)

                predicted=torch.max(y_val.data,1)[1]
                btach_corr=(predicted==y_test).sum()
                tst_corr+=btach_corr

            loss=loss.cpu().detach().numpy()
            test_losses.append(loss)
            test_correct.append(tst_corr)
            test_acc.append(tst_corr.item()/len(test_ds))

    model_fit_results = [train_losses, test_losses,train_correct, test_correct, train_acc, test_acc]        

    print(f'\nDuration: {time.time() - start_time:.0f} seconds')
    
    return model_fit_results
```

## CNN Model

```python
class MyCNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1=nn.Conv2d(3,6,3,1)
        self.conv2=nn.Conv2d(6,16,3,1)
        self.fc1=nn.Linear(16*54*54,120)
        self.fc2=nn.Linear(120,84)
        self.fc3=nn.Linear(84,20)
        self.fc4=nn.Linear(20,len(class_names))
        
    def forward(self,X):
        #print(X.shape)#torch.Size([10,3,224,224])
        X=F.relu(self.conv1(X))
        X=F.max_pool2d(X,2,2)
        X=F.relu(self.conv2(X))
        X=F.max_pool2d(X,2,2)
        X=X.view(-1,16*54*54)
        X=F.relu(self.fc1(X))
        X=F.relu(self.fc2(X))
        X=F.relu(self.fc3(X))
        X=self.fc4(X)
        X=F.log_softmax(X,dim=1)     
        #print(X.shape)#torch.Size([10,80])
        return X
```

```python
mycnn=MyCNN().to(device)
criterion=nn.CrossEntropyLoss()
optimizer=torch.optim.Adam(mycnn.parameters(),lr=0.001)
```

## Training function

```python
def gen_metrics_dict(metrics_list=None, metrics_dict=None, append_to=None, append_value=None):
    from collections import defaultdict
    metrics_dict = defaultdict(list)

    for key in metrics_list:
        metrics_dict[key] = []

    return metrics_dict
```

```python
def fit_train(batch_size, data_loader, model, optimizer, metrics_dict, epoch):
    corr = 0
    total_data_num = batch_size*len(data_loader)
    for batch_i, (X_data,Y_data) in enumerate(data_loader):
        
        batch_i+=1

        X_data, Y_data = X_data.to(device), Y_data.to(device)

        y_pred=model(X_data)
        loss=criterion(y_pred,Y_data)

        #true predictions
        predicted=torch.max(y_pred.data,1)[1]
        batch_corr=(predicted==Y_data).sum()
        corr+=batch_corr

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
        term = len(data_loader)//verbose_epoch_term
        if b%term==0:    # batch_size = 10, train_ds = 720 -> each batch : 72  ==> 72/36 =2번 출력
            print(f'epoch: {epoch+1:2}  batch: [{batch_size*b:4}/{total_data_num}]  loss: {loss.item():10.8f}  accuracy: {corr.item()*100/(batch_size*batch_i):7.3f}%')

    metrics_dict['train_loss'].append(loss.cpu().detach().numpy())
    metrics_dict['train_correct'].append(trn_corr)
    #-------- each Epoch accuracy -------------
    # accuracy = 100 * correct / len(trainset)
    # trainset, not train_loader
    # probably x in your case
    epoch_acc = corr.item()/total_data_num
    print('  ')
    print(f"Epoch: {epoch+1:2}  Accuracy :{epoch_acc*100:7.3f}%" ); print('-'*66);
    metrics_dict['train_accuracy'].append(epoch_acc)   
    
    return metrics_dict
```

```python
def fit_validation(batch_size, data_loader, model, optimizer, metrics_dict, epoch):
    corr = 0
    total_data_num = batch_size*len(data_loader)
    for batch_i, (X_data,Y_data) in enumerate(data_loader):
        batch_i+=1
        X_data, Y_data = X_data.to(device), Y_data.to(device)
        
        y_pred=model(X_data)
        loss=criterion(y_pred,Y_data)

        #true predictions
        predicted=torch.max(y_pred.data,1)[1]
        batch_corr=(predicted==Y_data).sum()
        corr+=batch_corr

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
    metrics_dict['test_loss'].append(loss.cpu().detach().numpy())
    metrics_dict['test_correct'].append(trn_corr)
    metrics_dict['test_acccuracy'].append(corr.item()/len(total_data_num))   
    
    return metrics_dict
```

```python
def Do_training(model, batch_size, epochs, matrics_list, verbose_epoch_term=2):
    
    import time
    start_time=time.time()
    
    history = gen_metrics_dict(matrics_list)
    
    print(f">>>> {model.__class__.__name__} Model fit start!!!"); print("  ")
    
    for epoch in range(epochs):

        history = fit_train((batch_size, data_loader, model, optimizer, metrics_dict, epoch, device)
                            
        with torch.no_grad():
            history = fit_validation((batch_size, data_loader, model, optimizer, metrics_dict=history, epoch, device)
                            

    model_fit_results = [train_losses, test_losses,train_correct, test_correct, train_acc, test_acc]        

    print(f'\nDuration: {time.time() - start_time:.0f} seconds')
    
    return model_fit_results
```
