# Food Torch CNN/AlexNet

```python
import os
import random
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")

from sklearn.model_selection import train_test_split

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader, random_split
from torchvision import datasets, transforms, models
from torchvision.utils import make_grid
```

```python
transform=transforms.Compose([
        transforms.RandomRotation(10),      # rotate +/- 10 degrees
        transforms.RandomHorizontalFlip(),  # reverse 50% of images
        transforms.Resize(224),             # resize shortest side to 224 pixels
        transforms.CenterCrop(224),         # crop longest side to 224 pixels at center
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406],
                             [0.229, 0.224, 0.225])
])
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
device
```

## Train and Test data Setting

```python
dataset=datasets.ImageFolder(root=("./imgs_data"),transform=transform)

class_names=dataset.classes
print('Class Num :',len(class_names))
print(class_names)
```

```python
torch.manual_seed(10)

dataset_size = len(dataset)
train_size = int(dataset_size * 0.8)
valid_size = int(dataset_size * 0.1)
test_size = dataset_size - train_size - valid_size

train_dataset, validation_dataset, test_dataset = random_split(dataset, [train_size, valid_size, test_size])

print(f"Training Data Size : {len(train_dataset)}")
print(f"Validation Data Size : {len(validation_dataset)}")
print(f"Testing Data Size : {len(test_dataset)}")
```

```python
batch_size = 10

train_loader=DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
valid_loader=DataLoader(validation_dataset, batch_size=batch_size, shuffle=True)
test_loader=DataLoader(test_dataset, batch_size=batch_size, shuffle=True)
```

## Training function

```python
def gen_history_dict(history_list):
    
    from collections import defaultdict
    history_dict = defaultdict(list)

    for key in history_list:
        history_dict[key] = []

    return history_dict
```

```python
def Do_training(model, epochs, historys, verbose_epoch_term=2):
    
    import time
    start_time=time.time()
    
    history = gen_history_dict(historys)
    
    print(f">>>> {model.__class__.__name__} Model fit start!!!"); print("  ")
    
    for epoch in range(epochs):

        trn_corr=0; val_corr=0;

        for b, (X_train,y_train) in enumerate(train_loader):
            b+=1
            X_train, y_train = X_train.to(device), y_train.to(device)

            y_pred=model(X_train)
            loss=criterion(y_pred,y_train)

            #true predictions
            predicted=torch.max(y_pred.data,1)[1]
            batch_corr=(predicted==y_train).sum()
            trn_corr+=batch_corr

            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            
            term = len(train_loader)//verbose_epoch_term
            if b%term==0:    # batch_size = 10, train_ds = 720 -> each batch : 72  ==> 72/36 =2번 출력
                print(f'epoch: {epoch+1:2}  batch: [{batch_size*b:4}/{len(train_dataset)}]  loss: {loss.item():10.8f}  accuracy: {trn_corr.item()*100/(batch_size*b):7.3f}%')

        #-------- each Epoch accuracy -------------
        # accuracy = 100 * correct / len(trainset)
        # trainset, not train_loader
        # probably x in your case
        epoch_acc = trn_corr.item()/len(train_dataset)
        history['train_accuracy'].append(epoch_acc)
        print('  ')
        print(f"Epoch: {epoch+1:2}  Accuracy :{epoch_acc*100:7.3f}%");
        history['train_loss'].append(loss.cpu().detach().numpy())
        history['train_correct'].append(trn_corr.cpu().detach().numpy())      

        with torch.no_grad():
            for b, (X_valid, y_valid) in enumerate(valid_loader):
                X_valid, y_valid = X_valid.to(device), y_valid.to(device)

                y_val_pred=model(X_valid)
                loss=criterion(y_val_pred, y_valid)

                predicted=torch.max(y_val_pred.data,1)[1]
                btach_corr=(predicted==y_valid).sum()
                val_corr+=btach_corr

            val_acc = val_corr.item()/len(validation_dataset)
            history['valid_accuracy'].append(val_acc)
            history['valid_loss'].append(loss.cpu().detach().numpy())
            history['valid_correct'].append(val_corr.cpu().detach().numpy())
            print(f"Epoch: {epoch+1:2}  Validation Accuracy :{val_acc*100:7.3f}%"); print('-'*66);

    print(f'\nDuration: {time.time() - start_time:.0f} seconds')
    
    return history
```

## CNN Model

```python
class MyCNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1=nn.Conv2d(3,6,3,1)
        self.conv2=nn.Conv2d(6,16,3,1)
        self.fc1=nn.Linear(16*54*54,120)
        self.fc2=nn.Linear(120,84)
        self.fc3=nn.Linear(84,20)
        self.fc4=nn.Linear(20,len(class_names))
        
    def forward(self,X):
        #print(X.shape)#torch.Size([10,3,224,224])
        X=F.relu(self.conv1(X))
        X=F.max_pool2d(X,2,2)
        X=F.relu(self.conv2(X))
        X=F.max_pool2d(X,2,2)
        X=X.view(-1,16*54*54)
        X=F.relu(self.fc1(X))
        X=F.relu(self.fc2(X))
        X=F.relu(self.fc3(X))
        X=self.fc4(X)
        X=F.log_softmax(X,dim=1)     
        #print(X.shape)#torch.Size([10,80])
        return X
```

```python
mycnn=MyCNN().to(device)
criterion=nn.CrossEntropyLoss()
optimizer=torch.optim.Adam(mycnn.parameters(),lr=0.001)
```

```python
mycnn
```

```python
def count_parameters(model):
    params = [p.numel() for p in model.parameters() if p.requires_grad]
    for item in params:
        print(f'{item:>8}')
    print(f'________\n{sum(params):>8}')
count_parameters(mycnn)
```

#### CNN fit

```python
target_hist = ['train_accuracy', 'train_loss', 'train_correct', 'valid_accuracy', 'valid_loss', 'valid_correct']

MyCNN_history = Do_training(model=mycnn, epochs=20, historys=target_hist, verbose_epoch_term=4)
```

## AlexNet Model

```python
alex= models.alexnet(pretrained=True)
alex
```

```python
# 불러온 모델은 학습 안시킨다.
# requires_grad=False 를 적용하면 모델의 특정 부분에 그라디언트 계산을 멈출 수 있습니다. 
# torch.no_grad() 와 가장 큰 차이는 requires_grad는 그라디언트를 저장은 한다는 것입니다. 
# 따라서 모델의 특정 부분은 freeze 하고 나머지는 학습시키는 등의 전략을 사용할 때 사용합니다.

for param in alex.parameters():
    param.requires_grad=False
```

```python
torch.manual_seed(10)

alex.classifier=nn.Sequential(nn.Linear(9216,1024),
                                     nn.ReLU(),
                                     nn.Dropout(p=0.5),
                                     nn.Linear(1024,80),
                                     nn.LogSoftmax(dim=1))
alex = alex.to(device)
alex
```

```python
count_parameters(alex)
```

```python
criterion=nn.CrossEntropyLoss()
optimizer=torch.optim.Adam(alex.classifier.parameters(),lr=0.001)
```

#### AlexNet fit

```python
target_hist = ['train_accuracy', 'train_loss', 'train_correct', 'valid_accuracy', 'valid_loss', 'valid_correct']

AlexNet_history = Do_training(model=alex, epochs=20, historys=target_hist, verbose_epoch_term=4)
```

## CNN Prediction

```python
mycnn.eval()
y_true=[]
y_pred=[]
with torch.no_grad():
    for test_data in test_loader:
        test_images, test_labels = test_data[0].to(device), test_data[1].to(device)
        pred = mycnn(test_images).argmax(dim=1)
        for i in range(len(pred)):
            y_true.append(test_labels[i].item())
            y_pred.append(pred[i].item())
print(y_pred[0:5])
```

```python
wrong_result = []

for n in range(0,len(y_true)):
    if y_true[n] != y_pred[n]:
        wrong_result.append(n)
        
print("Test Num :", len(y_true))
print("correct / wrong :", (len(y_true)-len(wrong_result)),'/', len(wrong_result))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_true,y_pred,target_names=class_names,digits=4)) # digits : 소수점 표시 자리
```
