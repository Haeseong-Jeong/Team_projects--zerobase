```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
mymodel = YOLO('../Data_total/best.pt')
mymodel.to('cuda')
```

```python
from glob import glob
#cutlet_path = '/content/drive/MyDrive/yogiyo_review//reviewer'
#cutlet_path = '../Data_total/reviews_data/reviewer'
cutlet_path = '../Data_total/new_JjamBbong/reviews_data/reviewer'
img_cnt = 0
wc = 0
a = 0

num = len(glob(cutlet_path + '*'))

for i in range(1,101):
    reviewer_imgs = glob(cutlet_path+str(i)+'/*')
    flag = False
    for img in reviewer_imgs:
        results = mymodel.predict(img, conf=0.5)
        box = results[0].boxes
        label = box.cls.cpu().detach().numpy()
        
        #if 13 in label:
        if (12 in label) or (13 in label) or (14 in label):
            flag = True
            pass
        else:
            #plots = results[0].plot()
            #plt.figure(figsize=(4,4))
            #plt.imshow(cv2.cvtColor(plots, cv2.COLOR_BGR2RGB))
            #plt.show()
            wc += 1
        img_cnt += 1
    if flag == False:
        a += 1
```
